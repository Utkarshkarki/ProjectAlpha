### So there are some missing values in the data which will be properly handeled in my data transformation components, i will use simple imputer for that purpose while creating a pipeline 

```python
df['occupation'].value_counts(normalize=True)
```

#### Data is distributed evenly i.e. it is not skewed so here we cannot use techniques like mode

```python
from sklearn.impute import SimpleImputer

# Create the imputer for constant value 'Unknown'
imputer = SimpleImputer(strategy='constant', fill_value='Unknown')

# Impute 'workclass' column
df['workclass'] = imputer.fit_transform(df[['workclass']]).ravel()

# Impute 'occupation' column
df['occupation'] = imputer.fit_transform(df[['occupation']]).ravel()

# Impute 'native.country' column
df['native.country'] = imputer.fit_transform(df[['native.country']]).ravel()
```

```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```
